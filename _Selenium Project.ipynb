# Q1 : Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. scrape first 10 jobs data(job-title, job-location, company_name, experience_required).

```python
#connecting to google chrome web driver
driver = webdriver.Chrome(r'C:\Users\Lenovo\Desktop\Worksheet\chromedriver.exe')
```

```python
# getting the webpage
driver.get("https://www.naukri.com/")
```

```python
#getting input field 'Skill, Designations, Companies' from web page storing it in a variable
role = driver.find_element(By.CLASS_NAME,"suggestor-input ")

#entering 'Data Analyst' in the input using 'role' variable
role.send_keys("Data Analyst")
```

```python
#getting “enter the location” field from web page storing it in a variable
city = driver.find_element(By.XPATH,'/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input')

#entering 'Bangalore' in the input using 'city' variable
city.send_keys("Bangalore")
```

```python
# locating search buttom from webpage & storing it in a variable 'button'
button = driver.find_element(By.XPATH,"/html/body/div[1]/div[6]/div/div/div[6]")

# clicking the search button.
button.click()
```

```python
#creating empty list to store the data
job_title_analyst = []
job_location_analyst = []
company_name_analyst = []
experience_required_analyst = []
```

```python
# lets use for loop to get first 10 jobs on the result page.

# finding job title & appending value to job_title list
titles = driver.find_elements(By.XPATH,'//a[@class="title fw500 ellipsis"]')
for i in titles[0:10]:
    title = i.text
    job_title_analyst.append(title)
    
    
# finding location & appending value to job_location list
locations = driver.find_elements(By.XPATH,'//li[@class="fleft grey-text br2 placeHolderLi location"]')
for i in locations[0:10]:
    location = i.text
    job_location_analyst.append(location)
    
    
# finding company name & appending value to company_name list
companies = driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
for i in companies[0:10]:
    company = i.text
    company_name_analyst.append(company)
    
    
# finding experience & appending value to company_name list
job_exp = driver.find_elements(By.XPATH,'//li[@class="fleft grey-text br2 placeHolderLi experience"]')
for i in job_exp[0:10]:
    exp = i.text
    experience_required_analyst.append(exp)
```

```python
# creating dataframe to store & visualize the data in a tabular format
ANALYST_JOBS = pd.DataFrame(list(zip(job_title_analyst, job_location_analyst, company_name_analyst, experience_required_analyst)),
                              columns =['Job_Title', 'Job_Location', 'Name_of_Company', 'Experience_Required'],
                              index=pd.RangeIndex(start=1, stop=(len(job_title_analyst)+1), name='Sr. No.'))


# Aligning the values of dataframe to the left
ANALYST_JOBS = ANALYST_JOBS.style.set_properties(**{'text-align': 'left'}).set_table_styles([ dict(selector='th',
                                                                                      props=[('text-align', 'left')])])


#showing the Dataframe
ANALYST_JOBS
```

```python
# closing login popup window
popup_close = driver.find_element(By.XPATH,"/html/body/div[2]/div/div/button")
popup_close.click()
```

```python
#getting search input field from web page storing it in a variable
input_field = driver.find_element(By.XPATH,"/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input")

#entering 'sunglasses' in the input using 'input_field' variable
input_field.send_keys("sunglasses")

# locating search button from webpage & storing it in a variable 'button'
button = driver.find_element(By.XPATH,"/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/button")

# clicking the search button.
button.click()
```

```python
# creating empty lists to store data
brand_list = []
product_description_list = []
price_with_discount_list = []
discount_list = []
```

```python
# lets use for loop to scape data within same website but different page numbers.

#dummy variable initialization for page count
first_page = 1 #starting page
last_page = 4 # page number to end scaping


for page in range(first_page,last_page):
    
    #to get brads
    brands = driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    for i in brands:
        brand_list.append(i.text)
    
    #to get product descriptions
    descriptions = driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    for i in descriptions:
        product_description_list.append(i.text)    
    
    #to get prices with discount
    prices = driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    for i in prices:
        price_with_discount_list.append(i.text)
    
    #to get total discount
    discounts = driver.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]')
    for i in discounts:
        discount_list.append(i.text)
        
    #getting the nextpage button & clicking next    
    next_page = driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')
    next_page.click()
    
    time.sleep(5) # to make selenium wait for 5 seconds to load the webpage properly..
```

```python
# Gathering only first 100 results & saving the data
brand_list = brand_list[0:100]
product_description_list = product_description_list[0:100]
price_with_discount_list = price_with_discount_list[0:100]
discount_list = discount_list[0:100]
```

# Q10: Write a python program to scrape the salary data from https://www.ambitionbox.com/ for Data Scientist designation. You have to scrape : 1. Company name 2. Number of salaries 3. Average salary 4. Minsalary 5. Max Salary

```python
#connecting to google chrome web driver
driver = webdriver.Chrome(r'C:\Users\Lenovo\Desktop\Worksheet\chromedriver.exe')
```

```python
#getting the webpage
driver.get("https://www.ambitionbox.com/")
```

```python
# closing login popup window
popup_close = driver.find_element(By.XPATH,'/html/body/div/div/div/div[3]/div/div/button/i')
popup_close.click()
```

```python
# Locating the salaries button & saving it in a variable
salary_button = driver.find_element(By.XPATH,'/html/body/div/div/div/div[1]/header/nav/ul/li[3]')

#Clicking on the salaries button
salary_button.click()
```

```python
# Locating the browse salaries button & saving it in a variable
browse_sal = driver.find_element(By.PARTIAL_LINK_TEXT, 'Salaries')

#Clicking on the browse salaries
browse_sal.click()
```

```python
#getting search input field from web page storing it in a variable
input_field = driver.find_element(By.XPATH,'/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/input')

# Entering “Data Scientist” in the search field
input_field.send_keys('Data Scientist')
```

```python
#locating the search button & saving it in a variable
button = driver.find_element(By.XPATH,'/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/div/div/div[1]/div/div/p')

#clicking on the search button
button.click()
```

```python
# creating empty list to store data
companies = []
total_rec = []
min_max = []
min_sal = []
max_sal = []
avg = []
exp = []



#to get company name
company = driver.find_elements(By.XPATH,'//a[@data-v-4c07f399]')
for i in company:
    companies.append(i.text.replace('\nData Scientist Salary'," ").strip()) 
    
    
#to get total salary records
sal_rec = driver.find_elements(By.XPATH,'//span[@class="datapoints"]')
for i in sal_rec:
    total_rec.append(i.text[9:12])
    
    
#to get min-max salary
tot_sal = driver.find_elements(By.XPATH,'//div[@class="salary-values"]')

for i in tot_sal:
    min_max.append(i.text.replace('\n'," to "))

for n in min_max:
    min_sal.append(n.split('to')[0].strip())
    max_sal.append(n.split('to')[1].strip())
    
#to get avg salary
avg_sal = driver.find_elements(By.XPATH,'//p[@class="averageCtc"]')
for i in avg_sal:
    avg.append(i.text)
    
    
#to get experience
expe = driver.find_elements(By.XPATH,'//div[@class="sbold-list-header"]')
for i in expe:
    exp.append(i.text.split('e')[0].strip())
```

```python
# creating dataframe to store & visualize the data in a tabular format
Sal_data = pd.DataFrame(list(zip(companies[0:10], total_rec, avg, min_sal, max_sal, exp)),
                              columns = ['Company','Total_Salary_Record', 'Average_Salary', 'Minimum_Salary', 'Maximum_Salary', 'Experience_Required'],
                              index = pd.RangeIndex(start=1, stop=(len(companies[0:10])+1), name='Sr. No.'))


# Aligning the values of dataframe to the left & showing the Dataframe
Sal_data.style.set_properties(**{'text-align': 'left'}).set_table_styles([ dict(selector='th',
                                                                                      props=[('text-align', 'left')])])
```
